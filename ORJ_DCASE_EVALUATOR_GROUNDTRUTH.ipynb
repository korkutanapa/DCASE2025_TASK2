```python
import os
import pandas as pd
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

EVAL_DIR = "/content/dcase2025_task2_evaluator"
PRED_DIR = "/content/dcase2025_task2_evaluator/teams/METU/TDA"

MACHINES = [
    "AutoTrash",
    "BandSealer",
    "CoffeeGrinder",
    "HomeCamera",
    "Polisher",
    "ScrewFeeder",
    "ToyPet",
    "ToyRCCar",
]

def read_no_header_csv(path, colname):
    df = pd.read_csv(path, header=None, names=["filename", colname])
    df["filename"] = df["filename"].astype(str)
    return df

for machine in MACHINES:
    print("\n" + "=" * 80)
    print(machine)
    print("=" * 80)

    gt_path = os.path.join(
        EVAL_DIR,
        "ground_truth_data",
        f"ground_truth_{machine}_section_00_test.csv"
    )

    domain_path = os.path.join(
        EVAL_DIR,
        "ground_truth_domain",
        f"ground_truth_{machine}_section_00_test.csv"
    )

    score_path = os.path.join(
        PRED_DIR,
        f"anomaly_score_{machine}_section_00_test.csv"
    )

    decision_path = os.path.join(
        PRED_DIR,
        f"decision_result_{machine}_section_00_test.csv"
    )

    gt = read_no_header_csv(gt_path, "y_true")
    domain = read_no_header_csv(domain_path, "domain")
    score = read_no_header_csv(score_path, "score")
    decision = read_no_header_csv(decision_path, "decision")

    df = (
        gt.merge(domain, on="filename", how="inner")
          .merge(score, on="filename", how="inner")
          .merge(decision, on="filename", how="inner")
    )

    print("Merged rows:", len(df))
    print("Ground truth rows:", len(gt))
    print("Score rows:", len(score))
    print("Decision rows:", len(decision))

    if len(df) != len(gt):
        print("WARNING: filename mismatch detected.")
        missing_pred = set(gt["filename"]) - set(score["filename"])
        extra_pred = set(score["filename"]) - set(gt["filename"])
        print("Missing prediction filenames:", list(sorted(missing_pred))[:10])
        print("Extra prediction filenames:", list(sorted(extra_pred))[:10])
        continue

    y_true = df["y_true"].astype(int).values
    y_score = df["score"].astype(float).values
    y_decision = df["decision"].astype(int).values

    print("\nALL FILES")
    print("AUC:", roc_auc_score(y_true, y_score))
    print(confusion_matrix(y_true, y_decision, labels=[0, 1]))

    for domain_name, domain_value in [("source", 0), ("target", 1)]:
        dfd = df[df["domain"].astype(int) == domain_value]

        y_true_d = dfd["y_true"].astype(int).values
        y_score_d = dfd["score"].astype(float).values
        y_decision_d = dfd["decision"].astype(int).values

        print(f"\n{domain_name.upper()} DECISION MATRIX")
        print("n:", len(dfd))
        print(confusion_matrix(y_true_d, y_decision_d, labels=[0, 1]))
        print("Accuracy:", accuracy_score(y_true_d, y_decision_d))
        print("Precision:", precision_score(y_true_d, y_decision_d, zero_division=0))
        print("Recall:", recall_score(y_true_d, y_decision_d, zero_division=0))
        print("F1:", f1_score(y_true_d, y_decision_d, zero_division=0))

    # Official-style domain AUC
    for domain_name, domain_value in [("source", 0), ("target", 1)]:
        mask_auc = (df["domain"].astype(int) == domain_value) | (df["y_true"].astype(int) != 0)

        y_true_auc = df.loc[mask_auc, "y_true"].astype(int).values
        y_score_auc = df.loc[mask_auc, "score"].astype(float).values

        print(f"\nOFFICIAL-STYLE {domain_name.upper()} AUC")
        print("n used:", len(y_true_auc))
        print("AUC:", roc_auc_score(y_true_auc, y_score_auc))
```
